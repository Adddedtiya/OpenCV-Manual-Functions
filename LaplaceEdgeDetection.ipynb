```python
import cv2 as cv
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
def showimg(image, title = ""):
    plt.imshow(image, cmap='gray')
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python
image_source = cv.imread('FlowerPatch.jpg', cv.IMREAD_COLOR)
image_source = cv.cvtColor(image_source, cv.COLOR_BGR2GRAY)
showimg(image_source, "Source Image")
```

```python
LaplaceKernel = np.matrix([
                 [-1, -1, -1],
                 [-1,  8, -1],
                 [-1, -1, -1]
])


LaplaceImage = cv.filter2D(image_source, -1, LaplaceKernel)
```

```python
showimg(LaplaceImage, "Laplacian Image")
```
