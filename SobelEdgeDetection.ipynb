```python
import cv2 as cv
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
def showimg(image, title = ""):
    plt.imshow(image, cmap='gray')
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python
image_source = cv.imread('FlowerPatch.jpg', cv.IMREAD_COLOR)
image_source = cv.cvtColor(image_source, cv.COLOR_BGR2GRAY)
showimg(image_source, "Source Image")
```

```python
SobelKernelX = np.matrix([
                 [1, 0, -1],
                 [2, 0, -2],
                 [1, 0, -1]
])

SobelKernelY = np.matrix([
                 [1,   2,  1],
                 [0,   0,  0],
                 [-1, -2, -1]
])



sobelX = cv.filter2D(image_source, -1, SobelKernelX)
sobelY = cv.filter2D(image_source, -1, SobelKernelY)

SobelCombined = sobelX + sobelY
```

```python
showimg(sobelX, "Image X")
print("")
showimg(sobelY, "Image Y")
print("")
showimg(SobelCombined, "Sobel Combined")
```
